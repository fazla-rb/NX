```python
import networkx as nx
import matplotlib.pyplot as plt

# Specify the path to your text file
txt_file = "nxz.txt"

# Create an empty NetworkX directed graph
G = nx.DiGraph()

# Read the edge list from the text file and add directed edges with weights and colors to the graph
with open(txt_file, 'r') as file:
    for line in file:
        # Split the line into parts
        parts = line.strip().split()
        if len(parts) >= 3:
            source, target, weight = parts[:3]
            # Convert the weight to a float (or the appropriate data type)
            weight = float(weight)
            # Create a dictionary to store edge attributes (weight and color)
            edge_attrs = {'weight': weight}
            if len(parts) >= 4:
                color = parts[3]
                edge_attrs['color'] = color
            G.add_edge(source, target, **edge_attrs)  # Add the edge with attributes

# Extract weights and colors for visualization
weights = [d['weight'] for s, t, d in G.edges(data=True)]
colors = [d.get('color', 'default_color') for s, t, d in G.edges(data=True)]

# Calculate node positions using a layout algorithm (e.g., spring_layout)
pos = nx.spring_layout(G)

# Draw the directed network graph with edge weights and colors
ax = plt.subplot(111)
nx.draw_networkx(
    G, pos, width=4, edge_color=colors, edge_cmap=plt.cm.Greys)
ax.margins(0.15, 0.15)

# Display the graph
plt.show()

```

```python
import networkx as nx

# Define your graph as a dictionary where keys are nodes, and values are lists of neighbors
graph = {
    'Winegroom': ['Uptown', 'Strawshop'],
    'Uptown': ['Strawshop', 'Amazelake','Winegroom'],
    'Strawshop': ['Winegroom'],
    'Province': ['Strawshop']
}

# Specify the filename where you want to save the adjacency list
output_file = 'example.adjlist'

# Open the file for writing and write the adjacency list
with open(output_file, 'w') as file:
    for node, neighbors in graph.items():
        # Write the node followed by its neighbors separated by spaces
        file.write(node + ' ' + ' '.join(neighbors) + '\n')

# Read the adjacency list
G = nx.read_adjlist(output_file)

# Add 'abbreviation' attribute to nodes
for v in G.nodes:
    G.nodes[v]['abbreviation'] = v[0]

# Write the graph to different formats
nx.write_edgelist(G, "out.edgelist")
nx.write_adjlist(G, "out.adjlist")
nx.write_gexf(G, "out.gexf")
```

Co occurance

```python
import networkx as nx

# Define your graph as a dictionary where keys are nodes, and values are lists of neighbors
graph = {
    'creature': ['justine'],
    'justine': ['victor', 'william', 'elizabeth'],
    'victor': ['william', 'elizabeth'],
    'william': ['victor', 'elizabeth'],
    'elizabeth': ['monster', 'henry'],
    'monster': ['henry'],
}

# Function to create a co-occurrence network from an existing graph
def co_occurrence_network_from_graph(input_graph):
    G = nx.Graph()

    # Initialize a dictionary to count co-occurrences
    edge_counts = {}

    # Iterate through the nodes in the input graph
    for node, neighbors in input_graph.items():
        # Add the node to the co-occurrence network
        G.add_node(node)

        # Iterate through the neighbors of the node
        for neighbor in neighbors:
            # Skip if the neighbor doesn't exist in the input graph
            if neighbor not in input_graph:
                continue

            # Add an edge between the node and its neighbor
            G.add_edge(node, neighbor)

            # Update the edge count
            edge = (node, neighbor)
            if edge in edge_counts:
                edge_counts[edge] += 1
            else:
                edge_counts[edge] = 1

    # Add edge counts as attributes to the edges
    for edge, count in edge_counts.items():
        G.edges[edge]['count'] = count

    return G

# Create the co-occurrence network from the provided graph
co_occurrence_net = co_occurrence_network_from_graph(graph)

# Sort the edges by 'count' in descending order and take the top 10 pairs
top_n_pairs = sorted(co_occurrence_net.edges(data=True), key=lambda e: e[2].get('count', 0), reverse=True)[:10]

# Print the top N pairs
for pair in top_n_pairs:
    print(pair)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Define your graph as a dictionary where keys are nodes, and values are lists of neighbors
graph = {
    'creature': ['justine'],
    'justine': ['victor', 'william', 'elizabeth'],
    'victor': ['william', 'elizabeth'],
    'william': ['victor', 'elizabeth'],
    'elizabeth': ['monster', 'henry'],
    'monster': ['henry'],
    'henry': ['monster'],
}

# Function to create a co-occurrence network from an existing graph
def co_occurrence_network_from_graph(input_graph):
    G = nx.Graph()

    # Initialize a dictionary to count co-occurrences
    edge_counts = {}

    # Iterate through the nodes in the input graph
    for node, neighbors in input_graph.items():
        # Add the node to the co-occurrence network
        G.add_node(node)

        # Iterate through the neighbors of the node
        for neighbor in neighbors:
            # Skip if the neighbor doesn't exist in the input graph
            if neighbor not in input_graph:
                continue

            # Add an edge between the node and its neighbor
            G.add_edge(node, neighbor)

            # Update the edge count
            edge = (node, neighbor)
            if edge in edge_counts:
                edge_counts[edge] += 1
            else:
                edge_counts[edge] = 1

    # Add edge counts as attributes to the edges
    for edge, count in edge_counts.items():
        G.edges[edge]['count'] = count

    return G

# Create the co-occurrence network from the provided graph
co_occurrence_net = co_occurrence_network_from_graph(graph)

# Define characters of interest
characters = [
    'creature', 'monster', 'victor', 'elizabeth',
    'william', 'henry', 'justine'
]

# Create a subgraph containing only the characters of interest
G_focus = co_occurrence_net.subgraph(characters)

# Create a list of edge counts
counts = [G_focus.edges[e]['count'] for e in G_focus.edges]

# Create spring layout
pos = nx.spring_layout(G_focus)

# Create figure and draw nodes
plt.figure(figsize=(10, 8))
nx.draw_networkx_nodes(G_focus, pos)

# Draw edges with edge color based on counts and alpha transparency
nx.draw_networkx_edges(
    G_focus, pos, width=8,
    edge_color=counts, edge_cmap=plt.cm.Blues, alpha=0.5
)
nx.draw_networkx_edges(G_focus, pos, edge_color="#7f7f7f", alpha=0.5)

# Draw labels
nx.draw_networkx_labels(G_focus, pos)

plt.tight_layout()

# Show the plot
plt.show()
```
